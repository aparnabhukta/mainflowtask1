# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Dataset 

```python
df = pd.read_csv(r"C:\Users\karri\Downloads\sales_data.csv")
```

```python
print(df.dtypes)
```

```python
print(df.head())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

# Line Plot

```python
sales_trend = df.groupby('Date')['Demand'].sum()
plt.figure(figsize=(12, 6))
plt.plot(sales_trend.index, sales_trend.values, marker='o', linestyle='-')
plt.title("Sales Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales (Demand)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Discount', y='Demand')
plt.title("Scatter Plot: Discount vs. Sales (Demand)")
plt.xlabel("Discount (%)")
plt.ylabel("Sales (Demand)")
plt.grid(True)
plt.show()
```

# Bar Chart

```python
region_sales = df.groupby('Region')['Demand'].sum().sort_values()
plt.figure(figsize=(8, 6))
region_sales.plot(kind='barh', color='skyblue')
plt.title("Total Sales by Region")
plt.xlabel("Total Sales (Demand)")
plt.ylabel("Region")
plt.grid(axis='x')
plt.show()
```

# Pie Chart

```python
category_sales = df.groupby('Category')['Demand'].sum()
plt.figure(figsize=(7, 7))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=140)
plt.title("Sales Distribution by Category")
plt.axis('equal')
plt.show()
```
